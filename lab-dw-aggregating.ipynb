This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by first performing data cleaning, formatting, and structuring.

```python
import pandas as pd

url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv"

df = pd.read_csv(url)


df.shape
```

```python
df.head()
```

```python
df.columns = df.columns.str.replace("ST", "state")

df.columns = (
    df.columns
      .str.lower()
      .str.replace(" ", "_")
)

df.columns
```

```python
gender_map = {
    "M": "M",
    "Male": "M",
    "male": "M",
    "F": "F",
    "Femal": "F",
    "Female": "F",
    "female": "F"
}

# Standardize gender column using map()
df['gender'] = df['gender'].replace(gender_map)

state_map = {
    "AZ": "Arizona",
    "Cali": "California",
    "WA": "Washington",
}

# Standardize gender column using map()
df['state'] = df['state'].replace(state_map)

df['education'] = df['education'].replace("Bachelors", "Bachelor")

# Remove '%' character from customer_lifetime_value
#df4['customer_lifetime_value'] = df4['customer_lifetime_value'].str.replace('%', '', regex=False)


vehicle_map = {
    "Sports Car": "Luxury",
    "Luxury SUV": "Luxury",
    "Luxury Car": "Luxury",
}

# Standardize gender column using map()
df['vehicle_class'] = df['vehicle_class'].replace(vehicle_map)

df.head()
```

```python
df.isnull().sum()

df = df.fillna(method="ffill")
```

```python
# Count number of policies sold by state and month
df['month'] = pd.to_datetime(df['effective_to_date']).dt.month_name()

policies_by_state_month = df.pivot_table(
    index='state',
    columns='month',
    values='policy',
    aggfunc='count',
    fill_value=0
)

print(policies_by_state_month)
```

5. The marketing team wants to analyze the number of policies sold by state and month. Present the data in a table where the months are arranged as columns and the states are arranged as rows.

```python
# Filter DataFrame for top 3 states
top_states = ["Arizona", "California", "Oregon"]

df_top_states = df[df['state'].isin(top_states)]

# Pivot table: number of policies sold by month for top 3 states
top_states_pivot = df_top_states.pivot_table(
    index='state',
    columns='month',
    values='policy',
    aggfunc='count',
)

print(top_states_pivot)
```
